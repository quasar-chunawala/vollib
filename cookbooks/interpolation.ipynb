```python
import sys
sys.path.insert(0, '/home/quantdev/repo/vollib/build/VolLib_Cpp')

from volpy.math import Points2D, LinearInterpolator1D

import numpy as np
import matplotlib.pyplot as plt

# Create a Points2D instance
data = Points2D([1.0, 2.0, 3.0], [2.0, 4.0, 6.0])

interpolator = LinearInterpolator1D
data.setInterpolator(interpolator)

knots = np.linspace(start=1.0, stop=3.0, num=200)
plt.xlim((1.0, 3.0))
plt.grid(True)
plt.xlabel(r"${x}$")
plt.ylabel(r"${y}$")
plt.title(r"Linear interpolation in $[1, 3]$")

yvals = [data.interpolate(x) for x in knots]

plt.plot(knots, yvals, 'b-', label=r'$\hat{y}$')

plt.show()
```
